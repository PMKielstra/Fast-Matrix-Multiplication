# Strassen Algorithm

This notebook implements the Strassen algorithm, which shows that $\omega \leq \log_2(7) \approx 2.8$.

```julia
function strassen_multiply(mul, A, B)
    p1 = mul(A[1, 1] + A[2, 2], B[1, 1] + B[2, 2])
    p2 = mul(A[2, 1] + A[2, 2], B[1, 1])
    p3 = mul(A[1, 1], B[1, 2] - B[2, 2])
    p4 = mul(A[2, 2], -B[1, 1] + B[2, 1])
    p5 = mul(A[1, 1] + A[1, 2], B[2, 2])
    p6 = mul(-A[1, 1] + A[2, 1], B[1, 1] + B[1, 2])
    p7 = mul(A[1, 2] - A[2, 2], B[2, 1] + B[2, 2])
    
    [(p1 + p4 - p5 + p7)       (p3 + p5)     ;
          (p2 + p4)       (p1 + p3 - p2 + p6)]
end
```

```julia
strassen = make_recursive_multiplier(strassen_multiply, (2, 2), (2, 2))
```
